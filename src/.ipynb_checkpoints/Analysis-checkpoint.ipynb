```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

analysisDF = pd.read_csv('analysis.csv',index_col=0,names=['# of vocabs','# of correct Answers','# of answered','Accuracy'])

analysisDF
```

```python
def plot_graph(name,x,y):
    y_keys=y.unique()
    if 0 not in y_keys:
        y_keys=np.append(y.unique(),0)
    y_keys=np.sort(y_keys)
    fig,ax = plt.subplots()
    ax.barh(x,y)
    for i, v in enumerate(y):
        ax.text(
            v
            ,i
            ,f"{v:1.2f}" if str(type(v)) == "<class 'float'>" else str(v)
            ,color='blue'
            ,fontweight='bold')
    if y_keys.max() > 100:
        ax.set_yticklabels(y_keys)
    ax.set_yticklabels(x)
    ax.set_title(name)
    fig.tight_layout()
    fig.set_figheight(20)
    fig.set_figwidth(40)

for column in analysisDF.columns:
    plot_graph(column,analysisDF.index,analysisDF[column])
```

A random baseline would yield an accuracy of around 25%. This is number is obtained by taking the probability of randomly picking the correct synonym given that each question has 4 options. Every model tested outperformed the random baseline by at least x0.88.  Furthermore, similar exercise was given out to the course's students. Collectively, they acheived an accuracy of 85.57% with a standard deviation of 13.18%. When comparing the various models' results, we can see that the three best models outperformed the humain gold-standard obtain from distributing the quiz to students. These are `fasttext-wiki-news-subwords-300`, `glove-wiki-gigaword-300` and `word2vec-google-news-300`. 

Notice, these three models all have the largest encoding size of 300. Furthermore, for every family of model tested, the encoding size had a positive correlation with the model's prediction accuracy. The fact that the worst performing model, which is `glove-twitter-25`, has the smallest embedding size falls in line with this observation. This is because an embedding size that is too small does not allow enough complexity for proper semantic representation. Interestingly, vocabulary size differences, when reaching very high limits, does not seem to have much of an impact on model performance.
