```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

analysisDF = pd.read_csv('analysis.csv',index_col=0,names=['# of vocabs','# of correct Answers','# of answered','Accuracy'])

analysisDF
```

```python
def plot_graph(name,x,y):
    y_keys=y.unique()
    if 0 not in y_keys:
        y_keys=np.append(y.unique(),0)
    y_keys=np.sort(y_keys)
    fig,ax = plt.subplots()
    ax.barh(x,y)
    for i, v in enumerate(y):
        ax.text(
            v
            ,i
            ,f"{v:1.2f}" if str(type(v)) == "<class 'float'>" else str(v)
            ,color='blue'
            ,fontweight='bold')
    if y_keys.max() > 100:
        ax.set_yticklabels(y_keys)
    ax.set_yticklabels(x)
    ax.set_title(name)
    fig.tight_layout()
    fig.set_figheight(20)
    fig.set_figwidth(40)

for column in analysisDF.columns:
    plot_graph(column,analysisDF.index,analysisDF[column])
```
